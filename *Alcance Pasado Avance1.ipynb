```python
from faker import Faker
import pandas as pd
import numpy as np

# Crear instancia de Faker
fake = Faker()

# Función para generar eventos de calendario
def generate_contextual_calendar_events(num_events):
    event_names = [
        "Reunión de estrategia con el equipo de ventas",
        "Revisión trimestral con clientes clave",
        "Taller de formación para nuevos empleados",
        "Presentación de resultados del proyecto",
        "Llamada con el proveedor"
    ]

    events = []
    for _ in range(num_events):
        event = {
            "event_name": np.random.choice(event_names),
            "date": fake.date_this_month(),
            "start_time": fake.time(),
            "duration": np.random.choice([30, 60, 90]),
            "participants": [fake.email() for _ in range(np.random.randint(2, 6))],
            "client_name": fake.company(),
            "classification": np.random.choice(["Urgente", "Informativo", "Regular"]),
        }
        events.append(event)
    return pd.DataFrame(events)

calendar_data = generate_contextual_calendar_events(500)
print(calendar_data.head())
```

```python
def generate_contextual_emails(num_emails):
    subjects = [
        "Reunión semanal de equipo",
        "Actualización del proyecto X",
        "Confirmación de agenda",
        "Solicitud urgente de aprobación",
        "Recordatorio de la entrega de reportes"
    ]
    bodies = [
        "Hola, adjunto la información solicitada para la reunión del lunes.",
        "Por favor, confirma tu asistencia al taller programado para el miércoles.",
        "Necesitamos tu feedback sobre el documento adjunto antes del viernes.",
        "El cliente ha solicitado cambios en el cronograma, revisemos los detalles.",
        "Recuerda completar el formulario de evaluación antes de final de mes."
    ]

    emails = []
    for _ in range(num_emails):
        email = {
            "subject": np.random.choice(subjects),
            "body": np.random.choice(bodies),
            "sender": fake.email(),
            "classification": np.random.choice(["Urgente", "Informativo", "Spam"]),
        }
        emails.append(email)
    return pd.DataFrame(emails)

email_data = generate_contextual_emails(1000)
print(email_data.head())
```

```python
def generate_travel_data(num_travels):
    travels = []
    for _ in range(num_travels):
        travel = {
            "destination": fake.city(),
            "date": fake.date_this_year(),
            "airline": fake.company(),
            "hotel": fake.company(),
            "price": round(np.random.uniform(100, 1000), 2),
        }
        travels.append(travel)
    return pd.DataFrame(travels)

travel_data = generate_travel_data(500)  # Generar 500 itinerarios
print(travel_data.head())
```

```python
def generate_contextual_documents(num_docs):
    titles = [
        "Contrato de Servicios",
        "Política de Privacidad",
        "Términos y Condiciones",
        "Acuerdo de Colaboración",
        "Manual de Usuario"
    ]
    errors = [
        "Falta de concordancia en fechas.",
        "Error en el formato del encabezado.",
        "Nombre del cliente incorrecto.",
        "Falta una cláusula obligatoria.",
        "Ortografía inconsistente en varias secciones."
    ]

    documents = []
    for _ in range(num_docs):
        doc = {
            "title": np.random.choice(titles),
            "content": fake.paragraph(nb_sentences=10),
            "errors": np.random.choice(errors, size=np.random.randint(1, 4), replace=False).tolist()
        }
        documents.append(doc)
    return pd.DataFrame(documents)

document_data = generate_contextual_documents(500)
print(document_data.head())

```

```python
def generate_bi_data(num_clients):
    bi_data = []
    for client_id in range(1, num_clients + 1):
        for month in range(1, 13):  # 12 meses
            data = {
                "client_id": client_id,
                "month": month,
                "sales": round(np.random.uniform(5000, 20000), 2),
                "web_traffic": np.random.randint(1000, 10000),
                "conversion_rate": round(np.random.uniform(0.01, 0.1), 2),
            }
            bi_data.append(data)
    return pd.DataFrame(bi_data)

bi_data = generate_bi_data(100)  # Generar datos de BI para 100 clientes
print(bi_data.head())
```

```python
# Valores faltantes por columna
print("Valores faltantes en Calendar Data:")
print(calendar_data.isnull().sum())

print("Valores faltantes en Email Data:")
print(email_data.isnull().sum())

print("Valores faltantes en Document Data:")
print(document_data.isnull().sum())

# Visualización de patrones de valores faltantes
msno.matrix(calendar_data)
plt.title("Valores Faltantes - Calendario")
plt.show()

msno.matrix(email_data)
plt.title("Valores Faltantes - Correos Electrónicos")
plt.show()

msno.matrix(document_data)
plt.title("Valores Faltantes - Documentos QA")
plt.show()
```

```python
# Estadísticas descriptivas para datos numéricos
print("Estadísticas descriptivas - Calendar Data")
print(calendar_data.describe())

print("Estadísticas descriptivas - Email Data")
print(email_data.describe())

print("Estadísticas descriptivas - Document Data")
print(document_data.describe())

# Frecuencias de variables categóricas
print("Frecuencia de clasificación en Calendar Data:")
print(calendar_data["classification"].value_counts())

print("Frecuencia de clasificación en Email Data:")
print(email_data["classification"].value_counts())
```

```python
# Cálculo del IQR para detectar valores atípicos en la duración de eventos
Q1 = calendar_data['duration'].quantile(0.25)
Q3 = calendar_data['duration'].quantile(0.75)
IQR = Q3 - Q1

# Rango permitido para valores no atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar valores atípicos
outliers = calendar_data[(calendar_data['duration'] < lower_bound) | (calendar_data['duration'] > upper_bound)]
print("Valores atípicos en duración de eventos:\n", outliers)
```
